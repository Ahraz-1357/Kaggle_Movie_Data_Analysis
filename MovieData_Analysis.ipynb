🎬 3. Movies per Year

🎬 4. Most Common Genres(study
)

```python
df['Genre'] = df['Genre'].str.split(', ')
genres = df.explode('Genre')

genre_count = genres['Genre'].value_counts().head(10)

sns.barplot(x=genre_count.values, y=genre_count.index)
plt.title('Top Genres')
plt.show()
```

🎬 5. Average Rating per Genre

```python
genre_rating = genres.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x=genre_rating.values, y=genre_rating.index)
plt.title('Avg Rating per Genre')
plt.show()
```

## YEAR vs Average Rating trend


```python
plt.figure(figsize=(25,8))
top10=df.groupby('Released_Year')['IMDB_Rating'].mean().sort_values(ascending=False).head(50)
sns.barplot(x=top10.index, y=top10.values)
plt.xlabel('Year')
plt.ylabel('Avg Rating')
plt.title("Year vs Average Rating trend")
plt.show()
```

## Runtime vs Rating

```python
df.head(5)
```

```python
df["RunTime"]= df['Runtime'].str.replace(' min','').astype(int)
sns.scatterplot(x='RunTime',y='IMDB_Rating',data=df)
plt.show()
```

```python
df["RunTime"] = df['Runtime'].str.replace(' min','').astype(int)

df['Runtime_Group'] = pd.cut(
    df['RunTime'],
    bins=[0,90,120,150,180,250],
    labels=['<90','90-120','120-150','150-180','180+']
)

sns.barplot(x='Runtime_Group', y='IMDB_Rating', data=df)
plt.show()
```

## 🎬 8. Top Directors (by Avg Rating)

```python
top_directors = df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top Directors')
plt.show()
```
